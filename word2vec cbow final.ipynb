```python
import pandas as pd
```

```python
data = pd.read_csv("D:\\New\\reviewdata\\reviewdata.csv", encoding="ISO-8859-1")
```

```python
data.head()
```

```python
import numpy as np
X = data.iloc[:,1:2].values.tolist()
```

```python
X
```

```python
for lst in X:
    for j, item in enumerate(lst):
        lst[j] = item.replace('\t', '')
```

```python
X
```

```python
Y = [' '.join(x) for x in X]
```

```python
Y
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
```

```python
words = [tokenizer.tokenize(sent) for sent in Y]
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
model = Word2Vec(words, size=2, min_count=1, seed=123456)
```

```python
words
```

```python
print('word vector for the word "hotel"')
print(model.wv['hotel'])
```

```python
print('Cosine distance between "hotel" and "people" is {}'.format(model.wv.distance('hotel','people')))
```

```python
import numpy as np
euc_dist = np.linalg.norm(model.wv['hotel'] - model.wv['people'])
print(euc_dist)
```

```python
print('Cosine similarity between "hotel" and "people" is {}'.format(model.wv.similarity('hotel','people')))
```

```python
w1 = ["hotel"]
model.wv.most_similar (positive=w1,topn=20)
```

```python
# save model
model.save('model.bin')
# load model
new_model = Word2Vec.load('model.bin')
print(new_model)
```
